```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.read_csv('GlobalWeatherRepository.csv')
df
```

```python
#countplot
sns.countplot(df.head(20),x="temperature_celsius",palette="Set1")
```

```python
sns.countplot(df.head(20),x="air_quality_PM2.5")
```

```python
sns.lineplot(df.head(10),x="last_updated",y="temperature_celsius")
plt.xticks(rotation=45)
```

```python
sns.lineplot(df.head(20),x="last_updated",y="condition_text")
plt.xticks(rotation=45)
```

```python
sns.barplot(df.tail(20),x="country",y="temperature_fahrenheit")
plt.xticks(rotation=45)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
top=df['condition_text'].value_counts().head(10)
plt.figure(figsize=(8,8))
plt.pie(top,labels=top.index,autopct='%1.1f%%',
        colors=sns.color_palette('Set2'))
plt.title("top 10 weather conditions")
plt.show()  
```

```python
plt.figure(figsize=(10,6))
sns.barplot(df.head(20),x="country",y="temperature_celsius",hue="condition_text",palette="Set1")
plt.xticks(rotation=45)
```

```python
plt.hist(df['humidity'], bins=20,color='pink',edgecolor='black')
plt.title("Humidity Distribution")
plt.xlabel("Humidity")
plt.ylabel("Frequency")
plt.show()
```

```python
import seaborn as sns
sns.heatmap(df[['temperature_celsius','humidity','wind_kph','pressure_mb']].corr(),annot=True,cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
sns.barplot(df.head(20),x="latitude",y="longitude",hue="condition_text")
plt.xticks(rotation=45)
plt.legend(title="Condition",bbox_to_anchor=(1.05, 1),loc="upper left")
```

```python
plt.hist(df['wind_kph'], bins=20)
plt.title("Wind Speed Distribution")
plt.xlabel("Wind Speed")
plt.ylabel("Frequency")
plt.show()
```

```python
df.groupby('country')['temperature_celsius'].mean().sort_values(ascending=False).head(10).plot(kind='bar')
plt.title("Top 10 Countries by Average Temperature")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.scatter(df.head(20)['temperature_celsius'],df.head(20)['humidity'])
plt.xlabel("Temperature")
plt.ylabel("Humidity")
plt.title("Temperature vs Humidity")
plt.show()
```
