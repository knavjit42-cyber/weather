```python
def category(temp):
    if temp < 15:
        return "cold"
    elif temp < 30:
        return "moderate"
    else:
        return "hot"

df["temp_category"] = df["temperature_celsius"].apply(category)
print(df.head())    
```

```python
df["temp_category"].value_counts().plot(kind="pie", autopct="%1.1f%%")
plt.title("temperature categories")
plt.ylabel("")
plt.show()
```
